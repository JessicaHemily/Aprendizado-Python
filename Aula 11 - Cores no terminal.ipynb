___
  **ANSI**

ESCAPE SEQUENCE

\033[_______________________ m
_____

**código do comportamento:**

\033[ STYLE m

**Código do texto:**

\033[ TEXT m

**Código da cor de fundo:**

\033[ BACK m

  ___
  **Utilizando as 3 opções**
   
    \033[style ; text ; back m

    STYLE
   
    0 None
   
    1 Bold
   
    4 Underline
   
    7 Negative
   
   
    TEXT                        BACKGROUND
 
    30      black       preto      40

    31      red         vermelho   41

    32      green       verde      42

    33      yellow      amarelo    43

    34      blue        azul       44

    35      Magenta     Magenta    45

    36      cyan        ciano      46

    37      grey        cinza      47

    97      white       branco     107

```python
print('\033[1;30;47mOlá mundo!')
```

```python
nome = 'Jessica'
print(f'Olá \033[31m{nome}\033[m!!')
```

**Criar uma variável cores (dicionário) para ajudar no decorrer do código:**

```python
cores = {'preto':'\033[30m',
         'vermelho':'\033[31m',
         'verde':'\033[32m',
         'semcor':'\033[m'}

print(f"Olá {cores['preto']} {nome} {cores['semcor']}")
print(f"Olá {cores['vermelho']} {nome} {cores['semcor']}")
print(f"Olá {cores['verde']} {nome} {cores['semcor']}")
```
